```python
import pandas as pd

from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    classification_report
)

from sklearn.model_selection import GridSearchCV
```

```python
train = pd.read_csv(r"C:\Users\veda\Downloads\churn-bigml-80.csv")
```

```python
test =  pd.read_csv(r"C:\Users\veda\Downloads\churn-bigml-20.csv")
```

```python
print(train.head())

print(train.info())
```

```python
le = LabelEncoder()

categorical_columns = [
    "State",
    "International plan",
    "Voice mail plan",
    "Churn"
]

for col in categorical_columns:
    train[col] = le.fit_transform(train[col])
    test[col] = le.transform(test[col])
```

```python
X_train = train.drop("Churn", axis=1)

y_train = train["Churn"]

X_test = test.drop("Churn", axis=1)

y_test = test["Churn"]
```

```python
lr = LogisticRegression(max_iter=1000)

lr.fit(X_train, y_train)

lr_pred = lr.predict(X_test)
```

```python
print("Logistic Regression")

print("Accuracy :", accuracy_score(y_test, lr_pred))

print("Precision :", precision_score(y_test, lr_pred))

print("Recall :", recall_score(y_test, lr_pred))

print("F1 Score :", f1_score(y_test, lr_pred))

print(classification_report(y_test, lr_pred))
```

```python
dt = DecisionTreeClassifier(random_state=42)

dt.fit(X_train, y_train)

dt_pred = dt.predict(X_test)
```

```python
print("Decision Tree")

print("Accuracy :", accuracy_score(y_test, dt_pred))

print("Precision :", precision_score(y_test, dt_pred))

print("Recall :", recall_score(y_test, dt_pred))

print("F1 Score :", f1_score(y_test, dt_pred))

print(classification_report(y_test, dt_pred))
```

```python
rf = RandomForestClassifier(random_state=42)

rf.fit(X_train, y_train)

rf_pred = rf.predict(X_test)
```

```python
print("Random Forest")

print("Accuracy :", accuracy_score(y_test, rf_pred))

print("Precision :", precision_score(y_test, rf_pred))

print("Recall :", recall_score(y_test, rf_pred))

print("F1 Score :", f1_score(y_test, rf_pred))

print(classification_report(y_test, rf_pred))
```

```python
param_grid = {
    "n_estimators": [50, 100],
    "max_depth": [5, 10, None]
}

grid = GridSearchCV(
    RandomForestClassifier(random_state=42),
    param_grid,
    cv=5,
    scoring="accuracy"
)

grid.fit(X_train, y_train)

print("Best Parameters:", grid.best_params_)

print("Best Accuracy:", grid.best_score_)
```

```python
best_model = grid.best_estimator_

prediction = best_model.predict(X_test)

print(classification_report(y_test, prediction))
```

```python
result = pd.DataFrame({
    "Actual": y_test,
    "Predicted": prediction
})

result.to_csv("classification_results.csv", index=False)

print(result.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

import nltk

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

from nltk.tokenize import word_tokenize

from textblob import TextBlob

from wordcloud import WordCloud
```

```python
nltk.download("punkt")
nltk.download("stopwords")
nltk.download("wordnet")
```

```python
df = pd.read_csv(r"C:\Users\veda\Downloads\3) Sentiment dataset.csv")
```

```python
print(df.head(1))
```

```python
df = df[["Text","Sentiment"]]
```

```python
df["Text"] = df["Text"].str.lower()
```

```python
df.dropna(inplace=True)
```

```python
import nltk

nltk.download('punkt_tab')
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
df["Tokens"] = df["Text"].apply(word_tokenize)
```

```python
stop_words = set(stopwords.words("english"))

df["Tokens"] = df["Tokens"].apply(
    lambda words: [word for word in words if word.isalpha() and word not in stop_words]
)
```

```python
lemmatizer = WordNetLemmatizer()

df["Tokens"] = df["Tokens"].apply(
    lambda words: [lemmatizer.lemmatize(word) for word in words]
)
```

```python
df["Clean_Text"] = df["Tokens"].apply(lambda words: " ".join(words))
```

```python
def get_sentiment(text):

    polarity = TextBlob(text).sentiment.polarity

    if polarity > 0:
        return "Positive"

    elif polarity < 0:
        return "Negative"

    else:
        return "Neutral"

df["Predicted_Sentiment"] = df["Clean_Text"].apply(get_sentiment)
```

```python
print(df[["Text","Predicted_Sentiment"]].head())
```

```python
sentiment_counts = df["Predicted_Sentiment"].value_counts()

plt.figure(figsize=(7,5))

plt.bar(sentiment_counts.index, sentiment_counts.values)

plt.title("Sentiment Distribution")

plt.xlabel("Sentiment")

plt.ylabel("Count")

plt.show()
```

```python
text = " ".join(df["Clean_Text"])

wordcloud = WordCloud(
    width=800,
    height=400,
    background_color="white"
).generate(text)

plt.figure(figsize=(12,6))

plt.imshow(wordcloud, interpolation="bilinear")

plt.axis("off")

plt.title("Word Cloud")

plt.show()
```
